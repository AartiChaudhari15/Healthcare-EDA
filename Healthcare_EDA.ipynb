```python
#import libraries which is required to perform task

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load the Dataset in (df) variable

df=pd.read_csv('/content/healthcare_dataset.csv')
df
```

```python
# Gives information the about the dataset

df.describe()
```

```python
#Rename column 'Medical Condition' as 'Diseases'

df = df.rename(columns={'Medical Condition' : 'Diseases'})
df.head()
```

```python
df.columns  #show columns which is available in dataset
```

```python
df.shape      #gives count of no.of columns and rows
```

**1) Count the Number of Patients per Doctor**

```python
doctor_patient_counts = df['Doctor'].value_counts()
print(doctor_patient_counts)
```

**2) Analyze The Distribution of Patients by Diseases**

```python
sns.countplot(x='Diseases', data=df)
plt.title('Distribution of Patients by Diseases')
```

**3) Analyze the distribution of patients by age and gender to understand the patient population**

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Patients by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Distribution of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**4) Examine the prevalence of different blood types among the patients**

```python
blood_types = df['Blood Type'].unique()
print('Unique Blood Types:', blood_types)

# Plot the prevalence of each blood type
plt.figure(figsize=(10, 6))
sns.countplot(x='Blood Type', data=df, order=blood_types, palette='Reds')
plt.title('Prevalence of Different Blood Types Among Patients')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.show()
```

**5) Analyze the frequency of admissions over time to identify any trends or patterns**

```python
# Convert 'Date of Admission' to datetime format
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])

# Extract year and month from the admission date
df['YearMonth'] = df['Date of Admission'].dt.to_period('M')

# Plot the frequency of admissions over time
plt.figure(figsize=(12, 6))
sns.countplot(x='YearMonth', data=df.sort_values('Date of Admission'), palette='viridis')
plt.title('Frequency of Admissions Over Time')
plt.xlabel('Year-Month of Admission')
plt.ylabel('Number of Admissions')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**6) Analyze the most prescribed medications**

```python
# Drop rows with missing values in the 'Medication' column
df_cleaned = df.dropna(subset=['Medication'])

# Count the occurrences of each medication
medication_counts = df_cleaned['Medication'].value_counts()

# Select the top N most prescribed medications
top_n = 5
top_medications = medication_counts.head(top_n)

# Plot the most prescribed medications
plt.figure(figsize=(12, 6))
sns.barplot(x=top_medications.index, y=top_medications.values, palette='rainbow')
plt.title('Top {} Most Prescribed Medications'.format(top_n))
plt.xlabel('Medication')
plt.ylabel('Prescription Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**7) Analyze billing amounts to identify the most expensive and least expensive treatments**

```python

# Drop rows with missing values in the 'Billing Amount' column
df_cleaned = df.dropna(subset=['Billing Amount'])

# Analyze the distribution of billing amounts
plt.figure(figsize=(12, 6))
sns.histplot(df_cleaned['Billing Amount'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Billing Amounts')
plt.xlabel('Billing Amount')
plt.ylabel('Count')
plt.show()

# Identify the most expensive and least expensive treatments
most_expensive = df_cleaned.loc[df_cleaned['Billing Amount'].idxmax()]
least_expensive = df_cleaned.loc[df_cleaned['Billing Amount'].idxmin()]

print('Most Expensive Treatment:')
print(most_expensive[['Medication', 'Billing Amount']])

print('\nLeast Expensive Treatment:')
print(least_expensive[['Medication', 'Billing Amount']])
```

**8) Examine the relationship between age, gender, and the prevalence of diseases**

```python
df_cleaned = df.dropna(subset=['Age', 'Gender', 'Diseases'])

# Create a violin plot to examine the relationship
plt.figure(figsize=(14, 8))
sns.violinplot(x='Gender', y='Age', hue='Diseases', data=df_cleaned, split=False, inner='quart', palette='muted')
plt.title('Relationship between Age, Gender, and Prevalence of Diseases')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
